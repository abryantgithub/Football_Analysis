# Reading FC Centre Back Analysis

- Taking League one Centre backs and then seeing where Reading FC Centre Backs rank amognst them

```python
import pandas as pd

# Importing data

df = pd.read_csv('league1_cbs.csv')
df.head(20)
```

```python
df.describe().T
```

```python
# Filter for 24/25 season
df_filtered = df[df['Season'] == '24/25'].copy()

# Create scaling factor for normalization (normalize to 46 appearances)
df_filtered['Scaling_Factor'] = 46 / df_filtered['Appearances']

# Normalize Goals
df_filtered['Goals_norm'] = df_filtered['Goals'] * df_filtered['Scaling_Factor']

# Create rankings for each stat
df_filtered['Rank_Rating'] = df_filtered['Rating'].rank(ascending=False)
df_filtered['Rank_Avg_Passes'] = df_filtered['Avg_Passes/90'].rank(ascending=False)
df_filtered['Rank_Key_Passes'] = df_filtered['Key_Passes/90'].rank(ascending=False)
df_filtered['Rank_Pass_Percentage'] = df_filtered['Pass_Percentage'].rank(ascending=False)
df_filtered['Rank_Tackles'] = df_filtered['Tackles/90'].rank(ascending=False)
df_filtered['Rank_Interceptions'] = df_filtered['Interceptions/90'].rank(ascending=False)
df_filtered['Rank_Fouls'] = df_filtered['Fouls/90'].rank(ascending=False)
df_filtered['Rank_Clearances'] = df_filtered['Clearances/90'].rank(ascending=False)
df_filtered['Rank_Blocks'] = df_filtered['Blocks/90'].rank(ascending=False)
df_filtered['Rank_Ground_Duels_Won'] = df_filtered['Ground_Duels_Won/90'].rank(ascending=False)
df_filtered['Rank_Aerial_Duels_Won'] = df_filtered['Aerial_duals_won/90'].rank(ascending=False)
df_filtered['Rank_Goals'] = df_filtered['Goals'].rank(ascending=False)

# Apply weighted scoring
df_filtered['Weighted_Score'] = (
    df_filtered['Rank_Rating'] * 9 +
    df_filtered['Rank_Tackles'] * 8 +
    df_filtered['Rank_Interceptions'] * 8 +
    df_filtered['Rank_Blocks'] * 7 +
    df_filtered['Rank_Ground_Duels_Won'] * 7 +
    df_filtered['Rank_Aerial_Duels_Won'] * 7 +
    df_filtered['Rank_Clearances'] * 6 +
    df_filtered['Rank_Fouls'] * 3 +
    df_filtered['Rank_Avg_Passes'] * 2
)

df_filtered['Overall_Rank'] = df_filtered['Weighted_Score'].rank(ascending=True)

# Sort by overall rank
df_final = df_filtered.sort_values('Overall_Rank')

# Output rankings â€” using the actual column names used above
print(df_final[['Name', 
                'Rating', 
                'Avg_Passes/90', 
                'Pass_Percentage',
                'Tackles/90',
                'Interceptions/90',
                'Goals_norm',
                'Fouls/90',
                'Clearances/90',
                'Blocks/90',
                'Ground_Duels_Won/90',
                'Aerial_duals_won/90',
                'Overall_Rank']])
```
